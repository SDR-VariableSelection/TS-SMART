```python
# %%
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
# Set global font properties
plt.rcParams['font.size'] = 14          # Set global font size
plt.rcParams['font.family'] = 'serif'   # Set global font family (e.g., 'serif', 'sans-serif', 'monospace')
plt.rcParams['font.serif'] = ['Times New Roman']  # Set specific serif font (if using 'serif' family)
plt.rcParams['axes.titlesize'] = 16     # Set font size for plot titles
plt.rcParams['axes.labelsize'] = 14     # Set font size for x and y labels
plt.rcParams['legend.fontsize'] = 12    # Set font size for legend
plt.rcParams['xtick.labelsize'] = 12    # Set font size for x-axis tick labels
plt.rcParams['ytick.labelsize'] = 12    # Set font size for y-axis tick labels
```

```python
## General loss
# %%
fname = f'simulation_005.pkl'
df = pd.read_pickle(fname, compression='gzip')  # load pkl back as dataframe
df = df.rename(columns={
    'loss_mddm': 'TS-MART',
    'loss_lasso': 'TS-SMART',
    'loss_group': 'TS-Group',
    'loss_sir': 'TS-SIR',
    'loss_rr': 'TS-RR',
    'example': 'Example',
    'size': 'Sample Size',
    'cov_str': "Sigma",
    # Add more columns to rename as needed
})
df_risk = pd.melt(df,
        id_vars=['Sample Size','Example','Sigma','seed'],
        value_vars=['TS-MART', 'TS-SMART', "TS-Group", "TS-SIR","TS-RR"],
        var_name='Method')
df_risk
df_risk.rename(columns={'value': 'General Loss'}, inplace=True)
```

```python
# %%
df2 = df_risk[(df_risk.Example < 3) & (df_risk.Sigma == 1)]
df2 = df2[(df2['Method'] != 'TS-Group') & (df2['Method'] != 'TS-SMART')]
palette_dict = {'TS-MART': '#d62728', 
                'TS-SMART': '#ff7f0e', 
                'TS-Group': '#9467bd', 
                'TS-SIR': '#1f77b4', 
                'TS-RR': '#2ca02c'}

g = sns.relplot(data=df2, x='Sample Size', y='General Loss',
                  col='Example', 
                  col_wrap=2,
                  hue='Method', 
                  kind='line',
                  palette=palette_dict,
                  errorbar = ('se',2),
                  legend= True)
g.legend.set_loc('upper right')
legend = g.legend
legend.set_title('')  # Change legend title fontsize
plt.show()
g.figure.savefig('Ex1-2_GL_Sig1.pdf', bbox_inches='tight')
```
